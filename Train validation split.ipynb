```python
import os
img_list = os.listdir("./images")
label_list = os.listdir("./labels")
```

```python
import numpy as np
names = np.array([tmp.split(".")[0] for tmp in img_list])
indices = np.random.choice(len(img_list),1296)
```

```python
test_names= names[indices]
test_names
```

```python
import os
import shutil
test_names= names[indices]
test_names = test_names.tolist()
img_dir = "./images/"
label_dir = "./labels/"
img_dest_dir = "../valid/images/"
label_dest_dir = "../valid/labels/"

for filename in test_names:
    try:
        shutil.move(img_dir+filename+".png", img_dest_dir)
        shutil.move(label_dir+filename+".txt", label_dest_dir)
    except:
        continue
```
